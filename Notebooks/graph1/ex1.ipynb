```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    message: str
```

```python
def comp_node(state : AgentState) -> AgentState:
    state["message"] = state["message"] + " you are doing amazing job !"
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("comp_node", comp_node)

graph.set_entry_point("comp_node")
graph.set_finish_point("comp_node")

app = graph.compile()
```

```python
result = app.invoke({"message" : "Bob"})

result["message"]
```
