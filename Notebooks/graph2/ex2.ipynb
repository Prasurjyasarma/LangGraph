```python
from typing import TypedDict, List
from langgraph.graph import StateGraph
```

```python
class agentState(TypedDict):
    values : List[int]
    name : str
    operation :str
    result : str
```

```python

def compute(state: agentState) -> agentState:

    if state["operation"] == "+":
        total = sum(state["values"])
        state["result"] = f"{state['name']}, you asked for sum, and the result is {total}"

    elif state["operation"] == "*":
        total = 1 
        for v in state["values"]:
            total *=v
        state["result"] = f"You asked for multiplication, and the result is {total}"

    else:
        state["result"] = f"Hi {state['name']}, what do you want?"

    return state
```

```python
graph = StateGraph(agentState)
graph.add_node("compute", compute)

graph.set_entry_point("compute")
graph.set_finish_point("compute")

app = graph.compile()
```

```python
result = app.invoke({
    "values" : [1,2,3,4],
    "name" : "Alice",
    "operation" : "*"
})

result["result"]
```
